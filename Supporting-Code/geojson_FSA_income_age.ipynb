```python
import requests
r = requests.get('https://sogramemon.github.io/plotly-javascript-challenge/data/Toronto2.geojson')
data = r.json()
```

```python
data['features']
```

```python
fsa_income = requests.get('http://127.0.0.1:5000/fsaIncomeAge').json()
```

```python
fsa_income_age_dictionary = {item['FSA'] : {'Avg_Age': item['Avg_Age'], 'avg_income': item['avg_income']} for item in fsa_income} #item['FSA']
```

```python
for item in data['features']:
    
    item['properties']['name'] = item['properties']['CFSAUID']
    item['properties']['Avg_Age'] = fsa_income_age_dictionary[item['properties']['CFSAUID']]['Avg_Age']
    item['properties']['avg_income'] = fsa_income_age_dictionary[item['properties']['CFSAUID']]['avg_income']
    
    if not fsa_income_age_dictionary[item['properties']['CFSAUID']]['Avg_Age']:
        print(f"Age is missing for {item['properties']['name']}")
        
    if not fsa_income_age_dictionary[item['properties']['CFSAUID']]['avg_income']:
        print(f"Income is missing for {item['properties']['name']}")
```

```python
from geojson import dump
with open('../Front-End/static/js/fsa_income_age.geojson', 'w') as f:
    dump(data, f)
```
