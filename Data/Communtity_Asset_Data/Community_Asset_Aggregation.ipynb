```python
# libraries
import pandas as pd
```

```python
#read Json file

df = pd.read_json("communityAssets.json")
```

```python
df.head()
```

```python
#find mean by no. of bedrooms and FSA
df_aggregate = df.groupby(["fsa", "category"])["agency_name"].count()
```

```python
df_aggregate.head()
```

```python
df_updated = df_aggregate.reset_index()
```

```python
df_updated.head()
```

```python
df_updated.rename(columns={"agency_name" : "no_assets"}, inplace=True)
```

```python
#save to file
df_updated.to_json("Community_assets_aggregate.json")
```
